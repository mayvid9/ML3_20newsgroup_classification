```python
features=[]    # selecting most common words from counter to form features
fil=open("C:\\Users\\Reliance Digital\\Desktop\\ml\\stopwords_en.txt",'r').read()
flist=fil.split()             # removing stop words from an external file
flist=[i.lower() for i in flist]
for i,j in count.most_common(10000):
    #print(type(i))
    if i.isalnum() and i not in string.digits and i not in flist:
        features.append(i)
```

```python
features
```

```python
feature=[]           # cleaning more non useful words from features
rem=['apr','gmt','make','fin','back','made','read','give','put','point','set','called','told','call','hard','means','left','told','works','hold','happened','gave','ii','na']
for i in features:
    if i not in rem:
        feature.append(i)
features=feature
len(features)
```

```python
dit={}            # forming dictionary for training and predicting 
f=0               # dict structure: {class_name:{file_name{words:count}}}
tot=0
for i in l:
    dit[i]={}
    tc=0
    tt=0
    dit[i]["total_words"]=0
    dit[i]['total_count']=0
    np="C:\\Users\\Reliance Digital\\Desktop\\ml\\20_newsgroups_train\\"+i
    for infile in glob.glob(os.path.join(np,'*')):
        g=[]
        n=infile.rfind('\\')
        fname=infile[n+2:]
        file=open(infile,'r').read()
        temp={}
        for t in features:
            temp[t]=0
        dit[i][fname]=temp
        g=file.split()
        g=[aa.lower() for aa in g]
        #tc+=len(g)
        count =Counter(g)
        tt+=1
        #tc=0
        for j in features:
            if j in count:
                dit[i][fname][j]=count[j]
                tc+=count[j]
        dit[i]["total_words"]=tc
    dit[i]['total_count']=tt
    f+=tc
    tot+=tt
dit["total_words"]=f
dit["total_count"]=tot
```

```python
def count(w1,clas):        # function to count total no of particular word w1 in a particular class
    s=0
    a=1
    for i in dit[clas]:
        for j in [dit[clas][i]]:
            if a<=2:
                a+=1
                continue
            else:
                if w1 not in dit[clas][i]:
                    return 0
                s+=dit[clas][i][w1]
    return s
#print(count('would','rec.sport.baseball'))
```

```python
def predict(dit,test):
    y_pred=[]
    for i in test:
        x_class=predict_single(dit,i)
        y_pred.append(x_class)
    return y_pred


def predict_single(dit,test):
    max_p=-1
    max_c='abc'
    che=True
    for cl in l:
        p=0
        for i in test:
            pro=(math.log(count(i,cl)+1)-math.log(dit[cl]["total_words"]+dit["total_words"]))
            p=p+pro
        p=p+(math.log(dit[cl]['total_count'])-math.log(dit['total_count']))
        if che==True:
            max_p=p
            max_c=cl
            che=False
        if che==False and max_p<p:
            max_p=p
            max_c=cl
    return max_c
```

```python
y=X_train.target
count_vect = CountVectorizer()
X_trai=count_vect.fit_transform(X_train.data)
clf=MultinomialNB()
clf.fit(X_trai,y) 
```

```python
test=datasets.fetch_20newsgroups(subset='test')
x_test=count_vect.transform(test.data)
clf.score(x_test,test.target)
```
